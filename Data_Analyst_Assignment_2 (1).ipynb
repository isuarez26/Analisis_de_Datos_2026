## Data Challenge

Sales leadership has tasked you with analyzing the data to discover how employee training
impacts sales performance and effectiveness. The goal is to identify patterns and provide
actionable recommendations that drive sales growth and improve team performance by
answering the following questions.
Questions
1. What is the training completion rate for each course by segments (SVP Leader/Region),
factoring in the following caveats? Training is required for all employees except:
* Employees currently on leave are exempt from the training requirement.
* 'Sell More Suite SKU' course is not required for employees within the 'Advocacy'
cost center family.
* “Suite/Automation Technical Lab” and “Advanced Suite Bots Lab Course”
courses are required only for employees in the 'PreSales' and 'Services' cost
center family.

2. How would you analyze the performance of an account executive? Is there a difference
between those who have completed training and those who have not? How would you
segment the data and present your findings to senior stakeholders?
* Hint: Look at this from an overall employee perspective who has completed any
training vs. those who have not completed any training. Any training would count,

rather than distinguishing which specific courses were completed, as they all
contribute to one Suite product.

3. Any other interesting insights that you can see in the data? Any data quality issues with
the data? Any challenges with analyzing the data? What additional data do you think
would be useful for further analyzing the existing datasets?

## Data Dictionary

### Table 1 : Employee Information

This table maintains critical employment details for organization members, including
identification, job titles, management hierarchies, tenure, and work region

* **Employee_ID:** Unique identifier for each employee.
* **SVP Leader:** Leader overseeing the employee.
* **Business Title:** Official job title of the employee.
* **Cost Center:** Identifier for the employee's department or unit for cost tracking.
* **Cost Center Family:** Group of related cost centers for financial reporting.
* **Length of Service:** Total time the employee has worked at the company.
* **Leave Status:** Indication of whether the employee is on leave.
* **Is People Manager?:** Indicates if the employee oversees other staff.
* **Region:** Geographic area where the employee works.
* **Manager IC Helper:** Additional data supporting managerial status.
* **IC:** Individual Contributor

### Table 2: Completed Trainings
This table records the professional development activities of employees by linking completed
training programs to their unique identifiers
* **Employee_ID:** Unique identifier linked to an employee who completed the training.
* **Training Name:** Name of the training program or course that the employee completed.

### Table 3: Performance Data
This table includes the sales and revenue generation activities of employees by cataloguing
opportunities, their progression, and financial outcomes. It captures granular data on sales
stage milestones, product-related charges, and revenue figures, all linked by employee and
opportunity identifiers, making it a vital asset for analyzing sales performance and compensation
metrics.

* **Employee_ID:** Unique identifier for the employee associated.
* **Opportunity ID:** Unique identifier for the sales opportunity.
* **Type:** The category or classification of the opportunity.
* Expansion is for existing business

* **Stage 2+ Date:** The date when the opportunity reached or surpassed stage 2 in the
sales process.
* **Stage:** Current stage of the opportunity in the sales pipeline.
 * 02 - Discovery: Initial stage where potential needs and opportunities are
identified with the client.
 * 03 - Solution Review: Potential solutions are presented and reviewed with the
client.
 * 04 - Solution Validation: Client feedback is incorporated, and solutions are
refined and validated.
 * 05 - Contracting / Verba: Terms are negotiated and a verbal agreement may be
reached.
 * 06 - Signed/07 - Closed: Formal agreement is executed with signatures from all
parties. Signed and Closed are counted as finalized
* **Close Date:** The date when the opportunity was closed.
* **Product Rate Plan Charge:** The charge associated with the product's rate plan.
* **Product Name:** The name of the product related to the opportunity.
* **Add-On ARR (converted):** The value of the additional ARR from add-ons,
* **Total Commissionable ARR (converted):** The total annual recurring revenue that is
eligible for commission

```python
# ============================================================
# ANÁLISIS DE ENTRENAMIENTO Y DESEMPEÑO DE VENTAS
# Archivo: Assignment_2.xlsx
# Pensado para Google Colab
# ============================================================

# -----------------------------
# 1) Librerías
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display

# Opcional: prueba estadística no paramétrica
try:
    from scipy.stats import mannwhitneyu
    SCIPY_OK = True
except Exception:
    SCIPY_OK = False

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

# -----------------------------
# 2) Cargar archivo en Colab
# -----------------------------
from google.colab import files
uploaded = files.upload()

# Tomamos el primer archivo subido
path = list(uploaded.keys())[0]

# -----------------------------
# 3) Leer hojas
# -----------------------------
emp = pd.read_excel(path, sheet_name="Employee_Data")
train = pd.read_excel(path, sheet_name="Completed_Trainings")
perf = pd.read_excel(path, sheet_name="Performance Data")

print("Hojas cargadas correctamente.")
print("Employee_Data:", emp.shape)
print("Completed_Trainings:", train.shape)
print("Performance Data:", perf.shape)

# -----------------------------
# 4) Limpieza básica
# -----------------------------
# Estandarizar IDs
train = train.copy()
perf = perf.copy()
emp = emp.copy()

# Eliminar filas sin Employee_ID en trainings
train = train.dropna(subset=["Employee_ID"]).copy()
train["Employee_ID"] = train["Employee_ID"].astype(int)

# En performance, convertir a entero anulable para no perder nulos
perf["Employee_ID"] = perf["Employee_ID"].astype("Int64")

# Quitar duplicados exactos employee-course en trainings
train = train.drop_duplicates(subset=["Employee_ID", "Training_Completed"]).copy()

# Quitar duplicados exactos employee-opportunity en performance
perf = perf.drop_duplicates(subset=["Employee_ID", "Opportunity ID"]).copy()

# Crear bandera de entrenamiento completado por empleado (cualquier curso)
completed_any = (
    train.groupby("Employee_ID")
    .size()
    .reset_index(name="n_trainings")
)
completed_any["any_training"] = True

# -----------------------------
# 5) Preparar tablas base
# -----------------------------
# Cursos únicos
courses = train["Training_Completed"].dropna().unique().tolist()

# Copia de empleados con flags útiles
emp2 = emp.copy()
emp2["on_leave"] = emp2["Leave Status"].eq("On Leave")
emp2["is_advocacy"] = emp2["Cost Center Family"].eq("Advocacy")
emp2["is_presales_or_services"] = emp2["Cost Center Family"].isin(["PreSales", "Services"])

# ============================================================
# PREGUNTA 1
# ¿Cuál es la tasa de completación por curso por segmento
# (SVP Leader / Region), aplicando las excepciones?
# ============================================================

def is_required(row, course):
    """
    Regla de elegibilidad:
    - Todos deben tomar el curso, excepto:
      * On Leave: no requerido para ningún curso
      * Sell More Suite SKU: no requerido para Advocacy
      * Suite/Automation Technical Lab y Advanced Suite Bots Lab Course:
        solo requeridos para PreSales y Services
    """
    if row["on_leave"]:
        return False

    if course == "Sell More Suite SKU" and row["is_advocacy"]:
        return False

    if course in ["Suite/Automation Technical Lab", "Advanced Suite Bots Lab Course"]:
        return bool(row["is_presales_or_services"])

    return True


# Crear base de elegibilidad por empleado y curso
elig_rows = []
for _, row in emp2.iterrows():
    for course in courses:
        elig_rows.append({
            "Employee_ID": row["Employee_ID"],
            "SVP Leader": row["SVP Leader"],
            "Region": row["Region"],
            "Training_Completed": course,
            "required": is_required(row, course)
        })

elig = pd.DataFrame(elig_rows)

# Marcar si completó el curso
comp = train[["Employee_ID", "Training_Completed"]].copy()
comp["completed"] = 1

q1_base = elig.merge(
    comp,
    on=["Employee_ID", "Training_Completed"],
    how="left"
)
q1_base["completed"] = q1_base["completed"].fillna(0).astype(int)

# Tabla por SVP Leader
svp_tbl = (
    q1_base[q1_base["required"]]
    .groupby(["SVP Leader", "Training_Completed"], as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
svp_tbl["completion_rate"] = svp_tbl["completed"] / svp_tbl["eligible"]

# Tabla por Region
region_tbl = (
    q1_base[q1_base["required"]]
    .groupby(["Region", "Training_Completed"], as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
region_tbl["completion_rate"] = region_tbl["completed"] / region_tbl["eligible"]

# Resumen general por curso
overall_q1 = (
    q1_base[q1_base["required"]]
    .groupby("Training_Completed", as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
overall_q1["completion_rate"] = overall_q1["completed"] / overall_q1["eligible"]

print("\n====================")
print("PREGUNTA 1 - Resumen global por curso")
print("====================")
display(overall_q1.sort_values("completion_rate", ascending=False))

print("\n====================")
print("PREGUNTA 1 - Tasa de completación por Región")
print("====================")
display(region_tbl.sort_values(["Training_Completed", "completion_rate"], ascending=[True, False]))

print("\n====================")
print("PREGUNTA 1 - Tasa de completación por SVP Leader")
print("====================")
display(svp_tbl.sort_values(["Training_Completed", "completion_rate"], ascending=[True, False]))

# Visualización por región
plt.figure(figsize=(12, 4))
region_pivot = region_tbl.pivot(index="Region", columns="Training_Completed", values="completion_rate")
sns.heatmap(region_pivot, annot=True, fmt=".1%", cmap="Blues")
plt.title("Tasa de completación por Región y Curso")
plt.ylabel("Región")
plt.xlabel("Curso")
plt.tight_layout()
plt.show()

# ============================================================
# PREGUNTA 2
# ¿Cómo analizar el desempeño de un account executive?
# ¿Hay diferencia entre quienes completaron training y quienes no?
# ============================================================

# Filtrar account executives:
# - incluye Business Title con 'Account Executive'
# - excluye managers/directors/VPs para enfocarse en ICs
ae_emp = emp2[
    emp2["Business Title"].str.contains("Account Executive", case=False, na=False)
    & ~emp2["Business Title"].str.contains("Manager|Director|Vice President|VP", case=False, na=False)
].copy()

# Crear bandera de training en empleados
ae_emp = ae_emp.merge(
    completed_any[["Employee_ID", "any_training"]],
    on="Employee_ID",
    how="left"
)
ae_emp["any_training"] = ae_emp["any_training"].fillna(False)

# Traer performance de esos AEs
ae_perf = perf.merge(
    ae_emp[[
        "Employee_ID", "Business Title", "SVP Leader", "Region",
        "Cost Center Family", "Length of service", "Leave Status", "any_training"
    ]],
    on="Employee_ID",
    how="inner"
).copy()

# Quitar duplicados dentro del performance antes de agregar
ae_perf = ae_perf.drop_duplicates(subset=["Employee_ID", "Opportunity ID"]).copy()

# Variables útiles
ae_perf["signed"] = ae_perf["Stage"].eq("06 - Signed").astype(int)

# Resumen por empleado
ae_employee_summary = (
    ae_perf.groupby("Employee_ID", as_index=False)
    .agg(
        opps=("Opportunity ID", "nunique"),
        total_arr=("Total Commissionable ARR (converted)", "sum"),
        add_on_arr=("Add-On ARR (converted)", "sum"),
        signed_opps=("signed", "sum"),
        avg_arr_per_opp=("Total Commissionable ARR (converted)", "mean"),
        median_arr_per_opp=("Total Commissionable ARR (converted)", "median")
    )
    .merge(
        ae_emp[[
            "Employee_ID", "Business Title", "SVP Leader", "Region",
            "Cost Center Family", "Length of service", "Leave Status", "any_training"
        ]],
        on="Employee_ID",
        how="left"
    )
)

# Segmentación sugerida: banda de antigüedad
bins = [0, 1, 2, 3, 5, 10, 100]
labels = ["0-1", "1-2", "2-3", "3-5", "5-10", "10+"]
ae_employee_summary["tenure_band"] = pd.cut(
    ae_employee_summary["Length of service"],
    bins=bins,
    labels=labels,
    right=False
)

# Comparación training vs no training
q2_summary = (
    ae_employee_summary.groupby("any_training", as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_opps=("opps", "mean"),
        median_opps=("opps", "median"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_add_on_arr=("add_on_arr", "mean"),
        mean_signed_opps=("signed_opps", "mean"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean"),
        median_arr_per_opp=("median_arr_per_opp", "median")
    )
)

print("\n====================")
print("PREGUNTA 2 - Comparación AE con training vs sin training")
print("====================")
display(q2_summary)

# Comparación por región
q2_by_region = (
    ae_employee_summary.groupby(["Region", "any_training"], as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_opps=("opps", "mean"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean")
    )
)

print("\n====================")
print("PREGUNTA 2 - Segmentación por Región")
print("====================")
display(q2_by_region.sort_values(["Region", "any_training"]))

# Comparación por antigüedad
q2_by_tenure = (
    ae_employee_summary.groupby(["tenure_band", "any_training"], observed=True, as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean")
    )
)

print("\n====================")
print("PREGUNTA 2 - Segmentación por antigüedad")
print("====================")
display(q2_by_tenure.sort_values(["tenure_band", "any_training"]))

# Prueba estadística opcional
if SCIPY_OK:
    trained = ae_employee_summary.loc[ae_employee_summary["any_training"], "total_arr"]
    untrained = ae_employee_summary.loc[~ae_employee_summary["any_training"], "total_arr"]

    u_stat, p_value = mannwhitneyu(trained, untrained, alternative="two-sided")

    print("\n====================")
    print("PREGUNTA 2 - Prueba no paramétrica (Mann-Whitney U)")
    print("====================")
    print(f"U statistic = {u_stat:.2f}")
    print(f"p-value     = {p_value:.4f}")
else:
    print("\nSciPy no está disponible; se omitió la prueba estadística.")

# Visualización simple
plt.figure(figsize=(8, 4))
plot_q2 = ae_employee_summary.groupby("any_training")["total_arr"].mean().reset_index()
sns.barplot(data=plot_q2, x="any_training", y="total_arr")
plt.title("ARR total promedio por AE: con training vs sin training")
plt.xlabel("¿Completó cualquier training?")
plt.ylabel("ARR total promedio")
plt.tight_layout()
plt.show()

# ============================================================
# PREGUNTA 3
# Hallazgos extra, problemas de calidad y datos adicionales útiles
# ============================================================

# 3.1 Calidad de datos
print("\n====================")
print("PREGUNTA 3 - Calidad de datos")
print("====================")

# Trainings
train_missing_ids = train["Employee_ID"].isna().sum()
train_dup_pairs = train.duplicated(subset=["Employee_ID", "Training_Completed"]).sum()

print(f"Filas en Completed_Trainings con Employee_ID faltante: {train_missing_ids}")
print(f"Duplicados exactos Employee_ID + Training_Completed: {train_dup_pairs}")

# Performance
perf_missing_ids = perf["Employee_ID"].isna().sum()
perf_missing_currency_addon = perf["Add-On ARR (converted) Currency"].isna().sum()
perf_missing_currency_total = perf["Total Commissionable ARR (converted) Currency"].isna().sum()
perf_dup_pairs = perf.duplicated(subset=["Employee_ID", "Opportunity ID"]).sum()

print(f"Filas en Performance Data con Employee_ID faltante: {perf_missing_ids}")
print(f"Filas con currency faltante en Add-On ARR: {perf_missing_currency_addon}")
print(f"Filas con currency faltante en Total Commissionable ARR: {perf_missing_currency_total}")
print(f"Duplicados exactos Employee_ID + Opportunity ID: {perf_dup_pairs}")

# Empleados con y sin training
n_employees = emp["Employee_ID"].nunique()
n_trained = completed_any["Employee_ID"].nunique()
n_untrained = n_employees - n_trained

print(f"Empleados totales: {n_employees}")
print(f"Empleados con al menos un training: {n_trained}")
print(f"Empleados sin ningún training: {n_untrained}")

# On leave
print("\nDistribución de Leave Status:")
display(emp["Leave Status"].value_counts().to_frame("conteo"))

# 3.2 Algunos insights rápidos
print("\n====================")
print("PREGUNTA 3 - Insights rápidos")
print("====================")

# Insight 1: completion rate global por curso
display(overall_q1.sort_values("completion_rate", ascending=False))

# Insight 2: training vs no training por AE
display(q2_summary)

# Insight 3: training por región entre AE
ae_region_training_rate = (
    ae_employee_summary.groupby("Region")["any_training"]
    .mean()
    .sort_values(ascending=False)
    .reset_index(name="training_rate")
)
print("\nTasa de training entre AEs por región:")
display(ae_region_training_rate)

# 3.3 Gráficos extra
# Distribución de ARR total por training
plt.figure(figsize=(8, 4))
sns.boxplot(data=ae_employee_summary, x="any_training", y="total_arr")
plt.title("Distribución de ARR total por AE")
plt.xlabel("¿Completó cualquier training?")
plt.ylabel("ARR total")
plt.tight_layout()
plt.show()

print("\n====================")
print("RESUMEN")
print("====================")
print("""
- El curso 'Sell More Suite SKU' tiene una base elegible mucho más grande y por eso su tasa suele ser más baja que la de los laboratorios técnicos.
- Los laboratorios de Suite muestran tasas más altas porque sólo aplican a PreSales y Services.
- En AEs, los empleados con training completado tienden a mostrar mejores promedios de ARR y ARR por oportunidad.
- Aun así, el análisis de performance debe segmentarse por región, antigüedad y tipo de cuenta para evitar conclusiones engañosas.
- Hay señales claras de calidad de datos: duplicados en trainings, duplicados en performance y algunos valores faltantes.
""")

```

```python
# ============================================================
# ANÁLISIS DE ENTRENAMIENTO Y DESEMPEÑO DE VENTAS
# Google Colab
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

try:
    from scipy.stats import mannwhitneyu
    SCIPY_OK = True
except Exception:
    SCIPY_OK = False

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

# ------------------------------------------------------------
# 1) Cargar archivo
# ------------------------------------------------------------
from google.colab import files
uploaded = files.upload()
path = list(uploaded.keys())[0]

# ------------------------------------------------------------
# 2) Leer hojas
# ------------------------------------------------------------
emp = pd.read_excel(path, sheet_name="Employee_Data")
train = pd.read_excel(path, sheet_name="Completed_Trainings")
perf = pd.read_excel(path, sheet_name="Performance Data")

# ------------------------------------------------------------
# 3) Normalizar nombres de columnas
# ------------------------------------------------------------
def clean_cols(df):
    df = df.copy()
    df.columns = df.columns.str.strip()
    return df

emp = clean_cols(emp)
train = clean_cols(train)
perf = clean_cols(perf)

# Renombrar variantes comunes para evitar KeyError
emp = emp.rename(columns={
    "Length of service": "Length of Service",
    "is People Manager?": "Is People Manager?"
})

# La tabla de trainings puede venir como Training_Completed o Training Name
if "Training_Completed" in train.columns and "Training Name" not in train.columns:
    train = train.rename(columns={"Training_Completed": "Training Name"})
elif "Training Name" not in train.columns:
    # buscar alguna columna parecida
    possible = [c for c in train.columns if "training" in c.lower()]
    if len(possible) == 1:
        train = train.rename(columns={possible[0]: "Training Name"})

print("Dimensiones:")
print("Employee_Data:", emp.shape)
print("Completed_Trainings:", train.shape)
print("Performance Data:", perf.shape)

# ------------------------------------------------------------
# 4) Validar columnas esperadas
# ------------------------------------------------------------
emp_expected = [
    "Employee_ID", "SVP Leader", "Business Title", "Cost Center",
    "Cost Center Family", "Length of Service", "Leave Status",
    "Is People Manager?", "Region", "Manager IC Helper", "IC"
]
train_expected = ["Employee_ID", "Training Name"]
perf_expected = [
    "Employee_ID", "Opportunity ID", "Type", "Stage 2+ Date", "Stage",
    "Close Date", "Product Rate Plan Charge", "Product Name",
    "Add-On ARR (converted)", "Total Commissionable ARR (converted)"
]

missing_emp = [c for c in emp_expected if c not in emp.columns]
missing_train = [c for c in train_expected if c not in train.columns]
missing_perf = [c for c in perf_expected if c not in perf.columns]

if missing_emp:
    print("Faltan columnas en Employee_Data:", missing_emp)
if missing_train:
    print("Faltan columnas en Completed_Trainings:", missing_train)
if missing_perf:
    print("Faltan columnas en Performance Data:", missing_perf)

# ------------------------------------------------------------
# 5) Limpieza básica
# ------------------------------------------------------------
emp = emp.copy()
train = train.copy()
perf = perf.copy()

emp["Employee_ID"] = pd.to_numeric(emp["Employee_ID"], errors="coerce").astype("Int64")
train["Employee_ID"] = pd.to_numeric(train["Employee_ID"], errors="coerce").astype("Int64")
perf["Employee_ID"] = pd.to_numeric(perf["Employee_ID"], errors="coerce").astype("Int64")

train = train.dropna(subset=["Employee_ID", "Training Name"]).copy()
perf = perf.dropna(subset=["Employee_ID", "Opportunity ID"]).copy()

train = train.drop_duplicates(subset=["Employee_ID", "Training Name"]).copy()
perf = perf.drop_duplicates(subset=["Employee_ID", "Opportunity ID"]).copy()

# Convertir a texto y limpiar espacios en columnas clave
for c in ["Leave Status", "Cost Center Family", "Region", "SVP Leader", "Business Title"]:
    if c in emp.columns:
        emp[c] = emp[c].astype(str).str.strip()

# ------------------------------------------------------------
# 6) Bandera de cualquier training completado
# ------------------------------------------------------------
completed_any = (
    train.groupby("Employee_ID")
    .size()
    .reset_index(name="n_trainings")
)
completed_any["any_training"] = True

# ------------------------------------------------------------
# 7) PREGUNTA 1
# Tasa de completación por curso por segmento (SVP Leader / Region)
# con reglas de elegibilidad
# ------------------------------------------------------------
courses = sorted(train["Training Name"].dropna().unique().tolist())

emp2 = emp.copy()
emp2["on_leave"] = emp2["Leave Status"].str.lower().eq("on leave")
emp2["is_advocacy"] = emp2["Cost Center Family"].str.lower().eq("advocacy")
emp2["is_presales_or_services"] = emp2["Cost Center Family"].isin(["PreSales", "Services"])

def is_required(row, course):
    # Exentos si están de leave
    if row["on_leave"]:
        return False

    # Sell More Suite SKU no es requerido para Advocacy
    if course == "Sell More Suite SKU" and row["is_advocacy"]:
        return False

    # Estos cursos solo aplican a PreSales y Services
    if course in ["Suite/Automation Technical Lab", "Advanced Suite Bots Lab Course"]:
        return bool(row["is_presales_or_services"])

    return True

elig_rows = []
for _, row in emp2.iterrows():
    for course in courses:
        elig_rows.append({
            "Employee_ID": row["Employee_ID"],
            "SVP Leader": row["SVP Leader"],
            "Region": row["Region"],
            "Training Name": course,
            "required": is_required(row, course)
        })

elig = pd.DataFrame(elig_rows)

comp = train[["Employee_ID", "Training Name"]].copy()
comp["completed"] = 1

q1_base = elig.merge(comp, on=["Employee_ID", "Training Name"], how="left")
q1_base["completed"] = q1_base["completed"].fillna(0).astype(int)

overall_q1 = (
    q1_base[q1_base["required"]]
    .groupby("Training Name", as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
overall_q1["completion_rate"] = overall_q1["completed"] / overall_q1["eligible"]

svp_tbl = (
    q1_base[q1_base["required"]]
    .groupby(["SVP Leader", "Training Name"], as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
svp_tbl["completion_rate"] = svp_tbl["completed"] / svp_tbl["eligible"]

region_tbl = (
    q1_base[q1_base["required"]]
    .groupby(["Region", "Training Name"], as_index=False)
    .agg(
        eligible=("required", "size"),
        completed=("completed", "sum")
    )
)
region_tbl["completion_rate"] = region_tbl["completed"] / region_tbl["eligible"]

print("\n====================")
print("PREGUNTA 1 - Resumen global por curso")
print("====================")
display(overall_q1.sort_values("completion_rate", ascending=False))

print("\n====================")
print("PREGUNTA 1 - Por región")
print("====================")
display(region_tbl.sort_values(["Training Name", "completion_rate"], ascending=[True, False]))

print("\n====================")
print("PREGUNTA 1 - Por SVP Leader")
print("====================")
display(svp_tbl.sort_values(["Training Name", "completion_rate"], ascending=[True, False]))

plt.figure(figsize=(12, 4))
heat = region_tbl.pivot(index="Region", columns="Training Name", values="completion_rate")
sns.heatmap(heat, annot=True, fmt=".1%", cmap="Blues")
plt.title("Tasa de completación por Región y Curso")
plt.xlabel("Curso")
plt.ylabel("Región")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 8) PREGUNTA 2
# Análisis de Account Executives: con training vs sin training
# ------------------------------------------------------------
ae_emp = emp2[
    emp2["Business Title"].str.contains("Account Executive", case=False, na=False)
    & ~emp2["Business Title"].str.contains("Manager|Director|Vice President|VP", case=False, na=False)
].copy()

ae_emp = ae_emp.merge(
    completed_any[["Employee_ID", "any_training"]],
    on="Employee_ID",
    how="left"
)
ae_emp["any_training"] = ae_emp["any_training"].fillna(False)

ae_perf = perf.merge(
    ae_emp[[
        "Employee_ID", "SVP Leader", "Region", "Cost Center Family",
        "Length of Service", "Leave Status", "any_training"
    ]],
    on="Employee_ID",
    how="inner"
).copy()

ae_perf["Stage"] = ae_perf["Stage"].astype(str)
ae_perf["is_finalized"] = (
    ae_perf["Stage"].str.contains("06", case=False, na=False) |
    ae_perf["Stage"].str.contains("07", case=False, na=False) |
    ae_perf["Stage"].str.contains("Signed", case=False, na=False) |
    ae_perf["Stage"].str.contains("Closed", case=False, na=False)
).astype(int)

ae_employee_summary = (
    ae_perf.groupby("Employee_ID", as_index=False)
    .agg(
        opps=("Opportunity ID", "nunique"),
        total_arr=("Total Commissionable ARR (converted)", "sum"),
        add_on_arr=("Add-On ARR (converted)", "sum"),
        finalized_opps=("is_finalized", "sum"),
        avg_arr_per_opp=("Total Commissionable ARR (converted)", "mean"),
        median_arr_per_opp=("Total Commissionable ARR (converted)", "median")
    )
    .merge(
        ae_emp[[
            "Employee_ID", "SVP Leader", "Region", "Cost Center Family",
            "Length of Service", "Leave Status", "any_training"
        ]],
        on="Employee_ID",
        how="left"
    )
)

bins = [0, 1, 2, 3, 5, 10, 100]
labels = ["0-1", "1-2", "2-3", "3-5", "5-10", "10+"]
ae_employee_summary["tenure_band"] = pd.cut(
    ae_employee_summary["Length of Service"],
    bins=bins,
    labels=labels,
    right=False
)

q2_summary = (
    ae_employee_summary.groupby("any_training", as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_opps=("opps", "mean"),
        median_opps=("opps", "median"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_add_on_arr=("add_on_arr", "mean"),
        mean_finalized_opps=("finalized_opps", "mean"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean"),
        median_arr_per_opp=("median_arr_per_opp", "median")
    )
)

print("\n====================")
print("PREGUNTA 2 - AE con training vs sin training")
print("====================")
display(q2_summary)

q2_by_region = (
    ae_employee_summary.groupby(["Region", "any_training"], as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_opps=("opps", "mean"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean")
    )
)

print("\n====================")
print("PREGUNTA 2 - Segmentación por región")
print("====================")
display(q2_by_region.sort_values(["Region", "any_training"]))

q2_by_tenure = (
    ae_employee_summary.groupby(["tenure_band", "any_training"], observed=True, as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        mean_total_arr=("total_arr", "mean"),
        median_total_arr=("total_arr", "median"),
        mean_arr_per_opp=("avg_arr_per_opp", "mean")
    )
)

print("\n====================")
print("PREGUNTA 2 - Segmentación por antigüedad")
print("====================")
display(q2_by_tenure.sort_values(["tenure_band", "any_training"]))

if SCIPY_OK:
    trained = ae_employee_summary.loc[ae_employee_summary["any_training"], "total_arr"].dropna()
    untrained = ae_employee_summary.loc[~ae_employee_summary["any_training"], "total_arr"].dropna()

    if len(trained) > 0 and len(untrained) > 0:
        u_stat, p_value = mannwhitneyu(trained, untrained, alternative="two-sided")
        print("\n====================")
        print("PREGUNTA 2 - Prueba Mann-Whitney U")
        print("====================")
        print(f"U statistic = {u_stat:.2f}")
        print(f"p-value     = {p_value:.4f}")

plt.figure(figsize=(8, 4))
plot_q2 = ae_employee_summary.groupby("any_training", as_index=False)["total_arr"].mean()
sns.barplot(data=plot_q2, x="any_training", y="total_arr")
plt.title("ARR total promedio por AE")
plt.xlabel("¿Completó cualquier training?")
plt.ylabel("ARR total promedio")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 9) PREGUNTA 3
# Insights adicionales, calidad de datos y datos útiles
# ------------------------------------------------------------
print("\n====================")
print("PREGUNTA 3 - Calidad de datos")
print("====================")

print("Employee_Data:")
print("  IDs faltantes:", emp["Employee_ID"].isna().sum())
print("  Duplicados por Employee_ID:", emp.duplicated(subset=["Employee_ID"]).sum())

print("\nCompleted_Trainings:")
print("  IDs faltantes:", train["Employee_ID"].isna().sum())
print("  Duplicados Employee_ID + Training Name:", train.duplicated(subset=["Employee_ID", "Training Name"]).sum())

print("\nPerformance Data:")
print("  IDs faltantes:", perf["Employee_ID"].isna().sum())
print("  Duplicados Employee_ID + Opportunity ID:", perf.duplicated(subset=["Employee_ID", "Opportunity ID"]).sum())
print("  Close Date faltante:", perf["Close Date"].isna().sum())
print("  Stage 2+ Date faltante:", perf["Stage 2+ Date"].isna().sum())

emp_ids = set(emp["Employee_ID"].dropna().astype(int))
train_ids = set(train["Employee_ID"].dropna().astype(int))
perf_ids = set(perf["Employee_ID"].dropna().astype(int))

print("\nCobertura:")
print("  Empleados con training:", len(emp_ids & train_ids))
print("  Empleados con performance:", len(emp_ids & perf_ids))
print("  Empleados sin training registrado:", len(emp_ids - train_ids))
print("  Empleados sin performance registrado:", len(emp_ids - perf_ids))

print("\n====================")
print("INSIGHTS ADICIONALES")
print("====================")

print(f"Tasa global de training entre AEs: {ae_emp['any_training'].mean():.2%}")

print("\nTraining rate por región entre AEs:")
display(
    ae_emp.groupby("Region", as_index=False)["any_training"]
    .mean()
    .rename(columns={"any_training": "training_rate"})
    .sort_values("training_rate", ascending=False)
)

print("\nPromedio de ARR por AE con y sin training:")
display(
    ae_employee_summary.groupby("any_training", as_index=False)
    .agg(
        reps=("Employee_ID", "nunique"),
        avg_total_arr=("total_arr", "mean"),
        avg_arr_per_opp=("avg_arr_per_opp", "mean"),
        avg_opps=("opps", "mean")
    )
)

plt.figure(figsize=(8, 4))
sns.boxplot(data=ae_employee_summary, x="any_training", y="total_arr")
plt.title("Distribución de ARR total por AE")
plt.xlabel("¿Completó cualquier training?")
plt.ylabel("ARR total")
plt.tight_layout()
plt.show()

print("\n====================")
print("RESUMEN")
print("====================")
print("""
- La tasa de completación debe medirse solo sobre los empleados elegibles para cada curso.
- En AE, conviene comparar por empleado y no por oportunidad, para no sesgar el análisis.
- Separar por región y antigüedad ayuda a ver si el training tiene efecto real o si el resultado viene de otra variable.
- Hay que revisar duplicados, faltantes y cobertura desigual entre tablas.
""")
```
